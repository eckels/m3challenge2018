```python
import pandas as pd  
import numpy as np
import math
import matplotlib.pyplot as plt  
dataset = pd.read_csv('perperson.csv')

def graph(formula, x_range, title, plotx, ploty):  
    x = np.array(x_range)  
    y = eval(formula)
    dataset.plot(x=plotx, y=ploty, style='o')
    plt.plot(x, y)  
    plt.title(title)
    plt.xlabel('Income (in dollars)')  
    plt.ylabel('Money spent on food (in dollars)') 
    plt.show()
    
c = 16604.34337
a = 4.267953504
b = 1.9848536 * 10**(-5)
e = math.exp(1)

graph('c/(1 + a*e**(-b * x))', range(0, 350000), 'Household', 'plot1', 'plot1points')

q = 5766.728826
w = 2.009830121
r = 1.0758084 * (10**-5)

graph('q/(1 + w*e**(-r * x))', range(0, 500000), 'Amount of money spent on food per person in a household with varying incomes', 'plot2', 'plot2points')
```
